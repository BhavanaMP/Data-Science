```python
import numpy as np
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline 
#to show the plot right after the plt.plot.no need to use plt.show()
import seaborn as sns
#Ref: https://www.kaggle.com/code/alexisbcook/hello-seaborn
```

```python
fifa_data = pd.read_csv("E:/Learning/ML/Datasets/data/fifa/fifa.csv", index_col='Date', parse_dates=True)
```

#### Info:
    - index_col - we are using date as index label for each row just to show visualisation of other data quickly.The index of the data will now be DateTimeIndex instead of usual RangeIndex(starts from 0 to len-1)
    - parse_date
        - boolean. If True -> try parsing the index (which is our date column here)
        - list of int or names. e.g. If [1, 2, 3] -> try parsing columns 1, 2, 3 each as a separate date column.
        - list of lists. e.g.  If [[1, 3]] -> combine columns 1 and 3 and parse as a single date column.
        - dict, e.g. {'foo' : [1, 3]} -> parse columns 1, 3 as date and call result 'foo'
        -This basically tells the notebook to understand the each row label as a date (as opposed to a number or other text with a different meaning).

```python
#Bar plot is used to see the average delay of each airline individually in all the months.

#Lets consider the American Airlines
plt.figure(figsize=(14, 5))
plt.title("Average delay of American airlines by month")
sns.barplot(x=flight_delays.index, y=flight_delays['AA'])
plt.xlabel("Month")
plt.ylabel("Average delay in minutes")
```

#### Info:
    - x=flight_delays.index - This determines what to use on the horizontal axis. In this case, we have selected the column that indexes the rows (in this case, the column containing the months).You must select the indexing column with flight_data.index, and it is not possible to use flight_data['Month'] (which will return an error). This is because when we loaded the dataset, the "Month" column was used to index the rows. We always have to use this special notation to select the indexing column.
    
    - y=flight_delays['NK'] - This sets the column in the data that will be used to determine the height of each bar. In this case, we select the 'NK' column.

#### ----------------------------------- heatmap() --------------------------------------------
    Heatmap is used to findout the patterns or relations between the features/columns or in the data.Mainly used to findout the correlation between the features

```python
plt.figure(figsize=(8, 8))
plt.title("Average Arrival Delay for Each Airline, by Month")
sns.heatmap(data=flight_delays, annot=True) #annot=True - This ensures that the values for each cell appear on the chart. 
plt.xlabel("Airlines")
plt.ylabel("Months")
```

#### Observation:
    What patterns can we detect from heatmap? 
        For instance, if we look closely, the months toward the end of the year (especially months 9-11) appear relatively dark for all airlines. This suggests that airlines are better (on average) at keeping schedule during these months!

#### ----------------------------- Scatterplot -----------------------------------------------
    we use scatter plots to highlight the relationship between two continuous variables. WE can also show relationship between 3 variables when using scatter plot with hue(color coding).

```python
#Lets use the insurance data which has insurance charges based on age, bmi, smoking and so on factors
insurance_data = pd.read_csv("E:\Learning\ML\Datasets\data\insurance.csv")
```

```python
insurance_data.info()
```

```python
insurance_data.head()
```

```python
#Lets use scatter plot to visualise how th charges gets affected by bmi
sns.scatterplot(x=insurance_data.bmi, y=insurance_data.charges)
plt.title("Effect of BMI on insurance charges")
```

#### Observation:
    We can observe a positive relationshop between bmi and insurance charges.With the increase in bmi of a person, the insurance charges tend to increase as they could be at higher risk of a chronic disease

#### ---------------------------regplot()------------------------------

```python
#To double-check the strength of this positive relationship, we can add a regression line that best fits the data of bmi and charges
sns.regplot(x=insurance_data.bmi, y=insurance_data.charges)
```

#### -------------------------------- scatterplot with hue() ------------------------------------------------

```python
#Now lets visualise how the smoking along with bmi effects the insurance charges.
#We can color-code the points by 'smoker', and plot the other two columns ('bmi', 'charges') on the axes.

sns.scatterplot(x=insurance_data.bmi, y=insurance_data.charges, hue=insurance_data.smoker)
```

#### Observation:
    If we observe, people who are smokers and having high BMI tend to pay more insurance charges compared to people who has high bmi but non smokers.To further emphasize this fact, we can use the sns.lmplot command to add two regression lines, corresponding to smokers and nonsmokers.

#### ------------------------------------- lmplot() --------------------------------------------------------
    - Plots data and regression model fits across a FacetGrid.
     When thinking about how to assign variables to different facets, a general rule is that it makes sense to use 'hue' for the most important comparison.
    But, this function combines :func:regplot and :class:FacetGrid. It is intended as a convenient interface to fit regression models across conditional subsets of a dataset.
 

```python
#Lets plot a regression line on both the groups of smokers and non smokers to show the kind of relationship between charges and smoking,bmi

sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)

#When statistical values are estimated, seaborn will use bootstrapping to compute confidence intervals and draw error bars representing the uncertainty of the estimate.(shaded regions aroung reg line)
```

#### Observation:
    Regression line for smokers has a much steeper slope, relative to the line for nonsmokers

#### --------------------------------- categorical scatter plot - swarmplot() -----------------------------------------
    we can adapt the design of the scatter plot to feature a categorical variable (like "smoker") on one of the main axes

```python
sns.swarmplot(x=insurance_data.smoker, y=insurance_data.charges)
```

#### Observation:
    The customers who pay the most are smokers whereas the customers who pay the least are non-smokers

#### -------------------------------------- histogram -----------------------------------------
    Histogram is used to show the distributions.It will show the count or density for each feature

```python
iris_data = pd.read_csv("E:/Learning/ML/Datasets/data/Iris.csv", index_col='Id')
iris_data.head()
```

```python
iris_data.info()
```

```python
#We can create a histogram to see how petal length varies in iris flowers
sns.histplot(x=iris_data.PetalLengthCm)

#it basically shows the distirbution of petal length.We can see more than 40 flowers has petal length of 1.
```

```python
#we can also color code the histogram of the data as per the feature/column that splits the data.
#data= provides the name of the variable that we used to read in the data
#x= sets the name of column with the data we want to plot
#hue= sets the column we'll use to split the data into different histograms

sns.histplot(data=iris_data, x="PetalLengthCm", hue="Species")
plt.title("Distribution of Petal Lengths, by Species")
```

#### Observation:
    If we observe, all the flowers with petal length <=2 belong to species iris-setosa.This clearly tells that petal length is more likely enough to distinguish setosa flowers from the data.We can also see that petal length for versicolor and virginica are almost same and merged at some point.Hence it is clear we may need another distinguishable feature to identify those species

#### ------------------------------ Density plot - kdeplot()----------------------------------
    KDE plot is kernel density estimation plot which is smoothened version of histogram plot.Setting shade=True colors the area below the curve (and data= chooses the column we would like to plot).

```python
sns.kdeplot(x=iris_data.PetalLengthCm, shade=True)
```

```python
#Color coding the kde plot
sns.kdeplot(data=iris_data, x=iris_data.PetalLengthCm, hue=iris_data.Species, shade=True)
plt.title("Distribution of Petal Lengths, by Species")
```

```python
# 2d KDE plot
#we can also plot kde with 2 variables
sns.kdeplot(x=iris_data.PetalLengthCm, y=iris_data.PetalWidthCm)
```

```python
sns.kdeplot(x=iris_data.PetalLengthCm, y=iris_data.PetalWidthCm, shade=True, cbar=True)
```

#### Observation:
    The color-coding shows us how likely we are to see different combinations of sepal width and petal length, where darker parts of the figure are more likely.

```python
#------------------------jointplot-----------------------------
#Draw a plot of two variables with bivariate and univariate graphs.For 2d kde, we can also use jointplot
#joint plot also shows the distribution of both the features on the paraller to their axes
#It plots the joint distribution between two variables along with each variables marginal distribution

sns.jointplot(x=iris_data.PetalLengthCm, y=iris_data.SepalWidthCm, kind='kde', shade=True, hue=iris_data.Species)
```

#### --------------------------- pairplot() ------------------------------------------------

```python
#pairplot(), takes a broader view.Just like scatterplotmatrix(pandas.plotting.scatter_matrix)
#It shows joint and marginal distributions for all pairwise relationships and for each variable, respectively:

sns.pairplot(data=iris_data, hue="Species")
```

#### Info:
    Since it's not always easy to decide how to best tell the story behind the data, the chart types are divided into three broad categories to help with this.
    
    --------------------------------------------------------------------------------------------------------------    
    -- Trends: 
        A trend is defined as a pattern of change.
        
        - sns.lineplot - Line charts are best to show trends over a period of time, and multiple lines can be used to show trends in more than one group.
        
    --------------------------------------------------------------------------------------------------------------
    -- Relationship:
        There are many different chart types that you can use to understand relationships between variables in your data.
        
        - sns.barplot - Bar charts are useful for comparing quantities corresponding to different groups.
        - sns.heatmap - Heatmaps can be used to find color-coded patterns in tables of numbers.
        - sns.scatterplot - Scatter plots show the relationship between two continuous variables; if color-coded, we can also show the relationship with a third categorical variable.
        - sns.regplot - Including a regression line in the scatter plot makes it easier to see any linear relationship between two variables.
        - sns.lmplot - This command is useful for drawing multiple regression lines, if the scatter plot contains multiple, color-coded groups.
        -sns.swarmplot - Categorical scatter plots show the relationship between a continuous variable and a categorical variable.
        -sns.pairplot
    -----------------------------------------------------------------------------------------------------------------
    
    -- Distribution:
        We visualize distributions to show the possible values that we can expect to see in a variable, along with how likely they are.
        
        - sns.histplot - Histograms show the distribution of a single numerical variable.
        - sns.kdeplot - KDE plots (or 2D KDE plots) show an estimated, smooth distribution of a single numerical variable (or two numerical variables).
        - sns.jointplot - This command is useful for simultaneously displaying a 2D KDE plot with the corresponding KDE plots for each individual variable.
        - sns.displot

```python
#we can also use displot to get both histogram and kde together
sns.displot(data=iris_data, x="PetalLengthCm", kde=True)
```

```python
sns.displot(data=iris_data, x="PetalLengthCm", kde=True, hue="Species")
```
