### Numerical Approximations using finite differences
    When you implement back propagation you'll find that there's a test called gradient checking that can really help you make sure that your implementation of back prop is correct. Because sometimes you write all these equations and you're just not 100% sure if you've got all the details right in internal back propagation. 
    
    So in order to build up gradient checking to make sure that implementation is backprop is correct, let's first talk about how to numerically approximate computations of gradients.
    
    Note this numerical approximation follows the concept of backprop intution that we have seen in MLS course backpropagation where we nudge the parameters by some extent and observe how it effect the Loss. (If w increases by epsilon, the J increases by k times epsilon where k is the rate of change in w w.r.to cost J which is nothing but gradient dJ/dw, and also the traingle concept)
    
    So let's take the function f and plot it here and lets say f(θ) = θ^3. Here θ is the parameter of function f
    
    Let's start off with θ with some value. Let's say θ = 1. 
    
    Numerical approximation done in 2 ways. 
    
    Before that why numerical approximation?
       NUmerical approximation method is used to find the "approximate gradients". First, know that gradients calculated by our backpropagation algorithm is also called analytical gradients. This is implemented by our logic of backpropagation algorithm. As our goal is to find out our implementation and the gradient calculating in backpropaagtion is right or wrong, we need to check them. We use the approximate gradients calculted using numerical approximation technique and compare them with analytical gradients. If the difference between them is very less, then we can confirm that our gradient calculation is right.
       
        Numerical approximation is a way to find out the slope of the function f at given θ. This slope is nothing but the gradient at θ, df/dθ. The rate of how f changes with the change in its parameter θ.
        
        It is calcualated in 2 ways.
            1. Using One sided nudge. 
            2. Using 2 sided nudge. 

#### One sided numerical approximation
        
        Here, lets suppose we have the function f with parameter θ. Lets imagine the paramater theta is at 1 i.e θ=1. Our goal is to find out how the change in parameter θ changes the function f i.e df/dθ. This is nothing but gradient.
        
        To do that, we can use one sided nudge numerical approximation technique.
        
        In one sided nudge, we nudge the parameter θ to the right of some small amount ε epsilon. So now θ changes to θ + ε. As initial θ is 1, and lets say the ε nudge is 0.01 to the right, meaning ε = 0.01, 
        
        Because θ changes to θ + ε its new value will be 1 + 0.01 =  1.01.
        
        Now, what we do is, we draw a little traingle from original theta to updated theta. We compute height(opposite side) over width(adjacent side) to get the slope or derivate of the function w.r.to its parameter θ. This is nothing but getting gradient approximately at that θ, df/dθ. 

        
        If we observe the traingle, the height is f(θ+ε) - f(θ) and the width is ε. Now the derivative of the θ can be calculated approximately as,
       
                   df/dθ = f(θ+ε) - f(θ) / ε
               
           Lets substitute all the values in the above equation to get the derivative.
           We defined ε = 0.01 and θ = 1 initially
           We know that f(θ) = θ^3, as θ = 1, f(θ) = 1^3 = 1
           Likewise Now, f(θ+ε) = (θ+ε)^3 = (1.01)^3 = 1.030301
           Now, df/dθ = f(θ+ε) - f(θ) / ε  = 1.030301 - 1 / 0.01 = 0.030301 / 0.01 = 3.0301
           We got df / dθ = 3.0301 approximately.
           
           Lets see what's the actual analytical derivative value using calculus derivations
           As f(θ) = θ^3, then df/dθ = 3 θ^2 
                                     = 3 (1^2)
                                     = 3
               Actual derivative df/dθ = 3
               
          Using one sided test, the approximated derivative is 3.0301 and actual derivative is 3 which are almost close. The approximation error is 0.0301. 
       
       Turns out there is more sophisticated numerical approximation approach than this, that results is very less approximation error. That is 2 sided numerical approximation
        
       Note: For a right angle traingle, tanθ = opposite side / adjacent side. We know that tanθ is nothing but slope

#### Two sided numerical approximation    
    
    Here, instead of just nudging θ only to the right to get θ + ε, we're going to nudge it to the right and nudge it to the left to also get θ - ε. 
    
    So if θ is 1 and ε is 0.01, then right nudge makes θ as 1.01 and left nudge makes it 0.99.
    
    

    
    It turns out that rather than taking the little triangle as above one sided approximation, we can now compute the height over the width of a bigger traingle. You can get a much better estimate of the gradient if you take these 2 points, f(θ - ε) and f(θ + ε).



    So for technical reasons which I won't go into, the height / width of this bigger green triangle gives you a much better approximation to the derivative at θ. 
    
    
    And you saw it yourself, taking just this lower triangle in the upper right is as if you have two triangles, right? The one on the upper right and the other one on the lower left. And you're kind of taking both of them into account by using this bigger green triangle. 
    
    So rather than a one sided difference, you're taking a two sided difference. 
    
    So let's work out the math. This point here is f(θ + ε). This point here is f(θ - ε). So the height of this big green triangle is f(θ + ε) - f(θ - ε). And then the width, is 1 epsilon + 1 epsilon which is 2 epsilon. So the width of this green triangle is 2ε. So the height over the width is going to be first the height, that's f(θ + ε) - f(θ - ε) divided by the width 2ε 
                                    
                                    df/dθ =  f(θ + ε) - f(θ - ε) / 2ε

    And this should hopefully be close to g(θ). So plug in the values in the above equation, remember f(θ) = θ^3 and if θ = 1 and ε = 0.01 => θ + ε = 1.01 and θ - ε = 0.99
                                     f(θ + ε) = (1.01)^3
                                     f(θ - ε) = (0.99)^3
                                     
                                  df/dθ = f(θ + ε) - f(θ - ε) / 2ε
                                         =  (1.01)^3 - (0.99)^3 / 2*0.01
                                         = 0.060002 / 0.02
                                         = 3.0001
                                      
                   Thus we got the approximated value of the derivative df/dθ as 3.0001. 
                   
      Lets see whats the usual derivative value df/dθ using usually calculated derivative analytically
                  f = θ^3 => df/dθ = 3θ^2
                  As θ = 1 => df/dθ = 3*(1^2) = 3
                  
       So these two values(approximated and actual analytical derivative) are actually very close to each other. The approximation error is now 0.0001.
       
       Whereas when we've taken the one sided of difference just θ and  θ + ε we had gotten 3.0301 as the approximation derivative and so the approximation error was 0.03 rather than 0.0001. 
       
       So with this two sided difference way of approximating the derivative, you find that this is extremely close to 3 actual derivative. And so this gives you a much greater confidence that df/dθ is probably a correct implementation of the derivative of f.

    When you use this method for gradient checking and back propagation, this turns out to run twice as slow as you were to use a one-sided derivative. 
    
    It turns out that in practice I think it's worth it to use this 2 sided method because it's just much more accurate. 
    
    And it turns out that for a non zero value of epsilon, you can show that the error of this approximation is on the order of epsilon squared, and remember epsilon is a very small number. So if epsilon is 0.01 which it is here, then epsilon squared is 0.0001. Whereas in contrast if we were to use the one sided difference, then the error is on the order of epsilon. And again, when epsilon is a number less than 1, then epsilon is actually much bigger than epsilon squared which is why it is actually much less accurate approximation. Which is why, when doing gradient checking, we rather use this two-sided difference  rather than just one sided difference which is less accurate.

    So you've seen how by taking a two sided difference, you can numerically verify whether or  not for a function f(θ) that someone else gives you is a correct implementation of the derivative of a function f.

### Gradient check using numerical approximation of finite differences
    Gradient checking is a technique that's helps us to save tons of time, and help us to find bugs in implementations of back propagation. 
    
    We could use it to debug, or to verify that your implementation and backpropagation is correct.
    
    Neural network will have parameters, like W1, b1 .. WL, bL. To implement gradient checking, the first thing you should do is "take all your parameters and reshape them into a giant vector data". 
    
    
    So what you should do is take W of each layer which is a matrix, and reshape it into a vector. You gotta take all of these Ws and reshape them into vectors, and then concatenate all of these things, so that you have a giant vector θ. 
    
    We know that the cost function J being a function of the Ws and Bs J(W1, B1, W2, B2...), but now you would have the cost function J being just a function of θ i.e J(θ). 
    
    Next, with W and B ordered the same way, you can also take dW[1], db[1] and so on, and initiate them into big, giant vector dθ of the same dimension as θ. 
    
    So same as before, we shape dW[1] into a vector, db[1] is already a vector. We shape dW[L], all of the dW's which are matrices. Remember, dW1 has the same dimension as W1. db1 has the same dimension as b1. So the same sort of reshaping and concatenation operation, you can then reshape all of these derivatives into a giant vector dθ which has the same dimension as θ. 
    
    So the question is, now, is the dθ, the gradient or the slope of the cost function J? 
    
    So here's how you implement gradient checking, and often abbreviate gradient checking to "grad check". 
    
    So first we remember that J is now a function of the giant parameter, θ, right? i.e J(θ) So expands to J is a function of θ1, θ2, θ3, and so on i.e J(θ1, θ2,..).

    Whatever's the dimension of this giant parameter vector θ, to implement grad check, what you're going to do is implements a loop so that for each i, so for each component of θ, let's compute dθ_approx_i to b  using a two sided difference. 
    
    So we will take J(θ1, θ2,..θi) and we're going to nudge θi to add epsilon to this i.e just increase θi by ε => θi + ε, and keep everything else the same. 
    
    And because we're taking a two sided difference, we're going to do the same on the other side with θi, but now minus epsilon => θi - ε. And then all of the other elements of theta are left alone. 
    
    And then we'll take this, and we'll divide it by 2ε and this should be approximately equal to dθi
    
    Note that dθi is the derivative of the cost function J w.r.to θi => dθi = ∂J / ∂θi. 
    
    So what we are going to do is we're going to compute to this for every value of i. 
    
    And at the end, you now end up with two vectors. You end up with this dθ approx, and this is going to be the same dimension as dθ and both of these are in turn the same dimension as θ. 
    
    And what you want to do is, check if these vectors are approximately equal to each other. 
    
    How to check 2 vectors are reasonably close to each other?
    ------------------------------------------------
        By computing Euclidean distance between the 2 vectors. Euclidean distance inturn uses L2 norm of the difference vector. 
   
    To compute the distance between these two vectors, we compute dθ approx - dθ and just take the l2 norm of this difference(l2norm(a_vec) = sqrt(sum of sqaures(a_vec))).
    
    Notice there's no square on top, so this is the sum of squares of elements of the differences, and then you take a square root, as you get the Euclidean distance.
    
    And then just to normalize by the lengths of these vectors, divide by ||dθ approx|| + ||dθ|| which is just take the Euclidean lengths of these vectors. 
    
Euclidean Distance = $\frac{|| dθ approx - dθ ||_2} {||dθ approx||_2 + ||dθ||_2} $
    
    And the row for the denominator is just in case any of these vectors are really small or really large, your the denominator turns this formula into a ratio. 
    
    In practice, use epsilon equals maybe 10^-7. And with this range of epsilon, if you find that this formula i.e euclidean distance gives you a value like 10^-7 or smaller, then that means that your derivative approximation is very likely correct.
    
     If the distance is maybe on the range of 10^-5, I would take a careful look. Maybe this is okay. But I might double-check the components of this vector θ, and make sure that none of the components are too large. And if some of the components of this difference are very large, then maybe you have a bug somewhere. 
     
     And if the distance gives you value like 10^-3, then you have to be much more concerned as maybe there's a bug somewhere. 


     
     The takeaway is, you should really be getting values much smaller than values like 10^-3. If any bigger than 10^-3, then it would be quite concerned. We would be seriously worried that there might be a bug. You should then look at the individual components of θ to see if there's a specific value of i for which dθ across i is very different from dθi approx. And use that to try to track down whether or not some of your derivative computations might be incorrect. 
     
     And after some amounts of debugging, if the distance ends up being very small value, then you probably have a correct implementation.

And after debugging for a while, If you find that it passes grad check with a small value, then you can be much more confident that it's then correct.

### Gradient Check Implementation Tips
    Some practical tips or some notes on how to actually go about implementing grad check for your neural network. 
    
    1. Don't use grad check in training, use only to debug. 
    
        So what I mean is that, computing dθi approx, for all the values of i, this is a very slow computation. So to implement gradient descent, you'd use backprop to compute dθ and just use backprop to compute the derivative. And it's only when you're debugging that you would compute the dθapprox to make sure it's close to dθ. But once you've done that, then you would turn off the grad check, and don't run this during every iteration of gradient descent, because that's just much too slow. 
        
    2. If an algorithm fails grad check, look at the components, look at the individual components, and try to identify the bug. 
    
        So what I mean by that is if dθ approx is very far from dθ, what I would do is look at the different values of i to see which are the values of dθ approx that are really very different than the values of dθ. So for eg, if you find that the values of θ or dθ, they're very far off, all correspond to dbl for some layer or for some layers, but the components for dw are quite close. Remember, different components of θ correspond to different components of b and w. When you find this is the case, then maybe you find that the bug is in how you're computing db, the derivative with respect to parameters b. And similarly, vice versa, if you find that the values that are very far, the values from dθ approx that are very far from dθ, you find all those components came from dw or from dw in a certain layer, then that might help you hone in on the location of the bug. This doesn't always let you identify the bug right away, but sometimes it helps you give you some guesses about where to track down the bug.

    3. When doing grad check, remember your regularization term if you're using regularization. 
    
        So if your cost function is J(θ) = 1 / m Σ L(y', y) + λ/2m Σ wl^2. And you should have that dθ is gradient of J with respect to θ, including this regularization term. So just remember to include that term. 
        
    4. Grad check doesn't work with dropout
        
        Because in every iteration, dropout is randomly eliminating different subsets of the hidden units. There isn't an easy to compute cost function J that dropout is doing gradient descent on. It turns out that dropout can be viewed as optimizing some cost function J, but it's cost function J defined by summing over all exponentially large subsets of nodes they could eliminate in any iteration. So the cost function J is very difficult to compute, and you're just sampling the cost function every time you eliminate different random subsets in those we use dropout. So it's difficult to use grad check to double check your computation with dropouts. So what I usually do is implement grad check without dropout. So if you want, you can set keep-prob in dropout to be equal to 1.0. And then after debug, turn on dropout and hope that my implementation of dropout was correct. 
        There are some other things you could do, like fix the pattern of nodes dropped and verify that grad check for that pattern is correct, but in practice I don't usually do that. So my recommendation is turn off dropout, use grad check to double check that your algorithm is at least correct without dropout, and then turn on dropout. 
        
      
      Finally, this is a subtlety. It is not impossible, rarely happens, but it's not impossible that your implementation of gradient descent is correct when w and b are close to 0, at random initialization. But that as you run gradient descent and w and b become bigger, maybe your implementation of backprop is correct only when w and b is close to 0, but it gets more inaccurate when w and b become large. 
      So one thing you could do, I don't do this very often, but one thing you could do is run grad check at random initialization and then train the network for a while so that w and b have some time to wander away from 0, from your small random initial values. And then run grad check again after you've trained for some number of iterations.

#### Exercise 3 - gradient_check
To show that the backward_propagation() function is correctly computing the gradient ∂J/∂θ, let's implement gradient checking.

**Instructions:**

- First compute "gradapprox" using the formula above and a small value of ε. Here are the Steps to follow:
     - 1. $θ^+ = θ + ε$
     - 2. $θ^- = θ - ε$
     - 3. $J^+ = J(θ^+) = θ^+x$
     - 4. $J^- = J(θ^-) = θ^-x$
     - 5. gradapprox = $\frac{J^+ \;- \;J^-}{2ε}$
     
- Then compute the gradient using backward propagation, and store the result in a variable "grad"
- Finally, compute the relative difference between "gradapprox" and the "grad" using the following formula:

   difference =  $\frac{[|| gradapprox - grad ||]_2}{||gradapprox||_2 + ||grad||_2}$
 
 
    You will need 3 Steps to compute this formula:

    - 1. compute the numerator using np.linalg.norm(...). Euclidean distance by calculating L2 norm of the distance
    - 2. compute the denominator. You will need to call np.linalg.norm(...) twice.
    - 3. divide them.
    - If this difference is small (say less than $10^{-7}$), you can be quite confident that you have computed your gradient correctly. Otherwise, there may be a mistake in the gradient computation.

```python
def gradient_check(x, theta, epsilon=1e-7, print_msg=False):
    """
    Implement the gradient checking presented in Figure 1.
    
    Arguments:
    x -- a float input
    theta -- our parameter, a float as well
    epsilon -- tiny shift to the input to compute approximated gradient with formula(1)
    
    Returns:
    difference -- difference (2) between the approximated gradient and the backward propagation gradient. Float output
    """
    theta_plus = theta + epsilon
    theta_minus = theta - epsilon
    J_plus = theta_plus * x
    J_minus = theta_minus * x
    gradapprox = (J_plus - J_minus) / (2*epsilon)
    grad = backward_propagation(x , theta)
    numerator = np.linalg.norm(gradapprox - grad)
    denominator = np.linalg.norm(gradapprox) + np.linalg.norm(grad)
    difference = numerator / denominator
    if print_msg:
        if difference > 2e-7:
            print ("\033[93m" + "There is a mistake in the backward propagation! difference = " + str(difference) + "\033[0m")
        else:
            print ("\033[92m" + "Your backward propagation works perfectly fine! difference = " + str(difference) + "\033[0m")
    
    return difference
```

```python
#test the function
x, theta = 3, 4
difference = gradient_check(x, theta, print_msg=True)
```

### 5 - N-Dimensional Gradient Checking
    The following figure describes the forward and backward propagation of your fraud detection model.



Figure 2: Deep neural network. LINEAR -> RELU -> LINEAR -> RELU -> LINEAR -> SIGMOID
    
    Let's look at your implementations for forward propagation and backward propagation.

```python
def sigmoid(x):
    s = 1 / (1 + np.exp(-x))
    return s

def relu(x):
    s = np.maximum(0, x)
    return s

def forward_propagation_n(X, Y, parameters):
    """
    Implements the forward propagation (and computes the cost) presented in Figure 3.
    
    Arguments:
    X -- training set for m examples
    Y -- labels for m examples 
    parameters -- python dictionary containing your parameters "W1", "b1", "W2", "b2", "W3", "b3":
                    W1 -- weight matrix of shape (5, 4)
                    b1 -- bias vector of shape (5, 1)
                    W2 -- weight matrix of shape (3, 5)
                    b2 -- bias vector of shape (3, 1)
                    W3 -- weight matrix of shape (1, 3)
                    b3 -- bias vector of shape (1, 1)
    
    Returns:
    cost -- the cost function (logistic cost for m examples)
    cache -- a tuple with the intermediate values (Z1, A1, W1, b1, Z2, A2, W2, b2, Z3, A3, W3, b3)

    """
    
    # retrieve parameters
    m = X.shape[1]
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]
    W3 = parameters["W3"]
    b3 = parameters["b3"]

    # LINEAR -> RELU -> LINEAR -> RELU -> LINEAR -> SIGMOID
    Z1 = np.dot(W1, X) + b1
    A1 = relu(Z1)
    Z2 = np.dot(W2, A1) + b2
    A2 = relu(Z2)
    Z3 = np.dot(W3, A2) + b3
    A3 = sigmoid(Z3)

    # Cost also in forward prop
    log_probs = np.multiply(-np.log(A3), Y) + np.multiply(-np.log(1 - A3), 1 - Y)
    cost = 1. / m * np.sum(log_probs)
    
    cache = (Z1, A1, W1, b1, Z2, A2, W2, b2, Z3, A3, W3, b3)
    
    return cost, cache
```

```python
def backward_propagation_n(X, Y, cache):
    """
    Implement the backward propagation presented in figure 2.
    
    Arguments:
    X -- input datapoint, of shape (input size, 1)
    Y -- true "label"
    cache -- cache output from forward_propagation_n()
    
    Returns:
    gradients -- A dictionary with the gradients of the cost with respect to each parameter, activation and pre-activation variables.
    """
    
    m = X.shape[1]
    (Z1, A1, W1, b1, Z2, A2, W2, b2, Z3, A3, W3, b3) = cache
    
    dZ3 = A3 - Y
    dW3 = 1. / m * np.dot(dZ3, A2.T)
    db3 = 1. / m * np.sum(dZ3, axis=1, keepdims=True)
    
    dA2 = np.dot(W3.T, dZ3)
    dZ2 = np.multiply(dA2, np.int64(A2 > 0))
    dW2 = 1. / m * np.dot(dZ2, A1.T) * 2
#     dW2 = 1. / m * np.dot(dZ2, A1.T) #correct implementaion
    db2 = 1. / m * np.sum(dZ2, axis=1, keepdims=True)
    
    dA1 = np.dot(W2.T, dZ2)
    dZ1 = np.multiply(dA1, np.int64(A1 > 0))
    dW1 = 1. / m * np.dot(dZ1, X.T)
    db1 = 4. / m * np.sum(dZ1, axis=1, keepdims=True)
#     db1 = 1. / m * np.sum(dZ1, axis=1, keepdims=True) #correct implementation
    
    gradients = {"dZ3": dZ3, "dW3": dW3, "db3": db3,
                 "dA2": dA2, "dZ2": dZ2, "dW2": dW2, "db2": db2,
                 "dA1": dA1, "dZ1": dZ1, "dW1": dW1, "db1": db1}
    
    return gradients
```

**How does gradient checking work?**

    As in Section 3 and 4, you want to compare "gradapprox" to the gradient computed by backpropagation. The formula is still:
    
$\frac{∂J}{∂θ} = lim_{ε->0}\frac{J(θ + ε)\; - \;J(θ - ε)}{2ε}$

    However, θ is not a scalar anymore. It is a dictionary called "parameters". The function "dictionary_to_vector()" converts the "parameters" dictionary into a vector called "values", obtained by reshaping all parameters (W1, b1, W2, b2, W3, b3) into vectors and concatenating them.

    The inverse function is "vector_to_dictionary" which outputs back the "parameters" dictionary.
    


Figure 2: dictionary_to_vector() and vector_to_dictionary(). 

    You will need these functions in gradient_check_n(). The "gradients" dictionary has also been converted into a vector "grad" using gradients_to_vector(), so you don't need to worry about that.

    Now, for every single parameter in your vector, you will apply the same procedure as for the gradient_check exercise. You will store each gradient approximation in a vector gradapprox. If the check goes as expected, each value in this approximation must match the real gradient values stored in the grad vector.

    Note that grad is calculated using the function gradients_to_vector, which uses the gradients outputs of the backward_propagation_n function.

**Instructions: Here is pseudo-code that will help you implement the gradient check**

- For each i in num_parameters:

    - To compute J_plus[i]:
        - Set $θ^+$  to np.copy(parameters_values)
        - Set ${θ_i}^+$ to ${θ_i}^+ +  ε $ 
        - Calculate ${J_i}^+$  using to forward_propagation_n(x, y, vector_to_dictionary($θ^+$)).

    - To compute J_minus[i]: do the same thing with $θ^-$
    - Compute $gradapprox[i] = \frac{{J_i}^+ \;- \;{J_i}^-}{2ε}$
 

    Thus, you get a vector gradapprox, where gradapprox[i] is an approximation of the gradient with respect to parameter_values[i]. You can now compare this gradapprox vector to the gradients vector from backpropagation. Just like for the 1D case (Steps 1', 2', 3'), compute:
 
 difference =  $\frac{[|| gradapprox - grad ||]_2}{||gradapprox||_2 + ||grad||_2}$

    Note: Use np.linalg.norm to get the norms

```python
def dictionary_to_vector(parameters):
    """
    Roll all our parameters dictionary into a single vector(column vector) satisfying our specific required shape.
    """
    keys = []
    count = 0
    for key in ["W1", "b1", "W2", "b2", "W3", "b3"]:
        new_vector = np.reshape(parameters[key], (-1, 1)) #unrolling to a single vector
        keys = keys + [key] * new_vector.shape[0] #eg W1 * 4 = [W1 W1 W1 W1]
#         print(keys)
        if count == 0:
            theta = new_vector
        else:
            theta = np.concatenate((theta, new_vector), axis=0) #concatenation vertically
        count = count + 1
    return theta, keys
    

def gradients_to_vector(grads):
    """
    Roll all our gradients dictionary into a single vector(column vector) satisfying our specific required shape.
    """
    keys = []
    count = 0
    for key in ["dW1", "db1", "dW2", "db2", "dW3", "db3"]:
        new_vector = np.reshape(grads[key], (-1, 1))
        
        if count == 0:
            dtheta = new_vector
        else:
            dtheta = np.concatenate((dtheta, new_vector), axis=0)
        count = count + 1
    return dtheta
        

def vector_to_dictionary(theta):
    """
    Unroll all our parameters dictionary from a single vector satisfying our specific required shape.
    """
    params = {}
#     print(theta.shape) #(47, 1)
    params["W1"] = theta[: 20].reshape((5, 4))
    params["b1"] = theta[20 : 25].reshape((5, 1))
    params["W2"] = theta[25 : 40].reshape((3, 5))
    params["b2"] = theta[40 : 43].reshape((3, 1))
    params["W3"] = theta[43 : 46].reshape((1, 3))
    params["b3"] = theta[46 : 47].reshape((1, 1))
    return params
    
def gradient_check_n(parameters, grads, X, Y, epsilon=1e-7, print_msg=False):
    """
    Checks if backward_propagation_n computes correctly the gradient of the cost output by forward_propagation_n
    
    Arguments:
    parameters -- python dictionary containing your parameters "W1", "b1", "W2", "b2", "W3", "b3"
    grad -- output of backward_propagation_n, contains gradients of the cost with respect to the parameters 
    X -- input datapoint, of shape (input size, number of examples)
    Y -- true "label"
    epsilon -- tiny shift to the input to compute approximated gradient with formula(1)
    
    Returns:
    difference -- difference (2) between the approximated gradient and the backward propagation gradient
    """
    parameter_values, _ = dictionary_to_vector(parameters)
#     print(parameter_values.shape) #(47, 1)
    grad = gradients_to_vector(grads)
#     print(grad.shape) #(47, 1)
    num_parameters = parameter_values.shape[0]
    J_plus = np.zeros((num_parameters, 1)) #for each parameter increase we gonna calculate cost
    J_minus = np.zeros((num_parameters, 1))
    gradapprox = np.zeros((num_parameters, 1))
    
    for i in range(num_parameters):
        theta_plus = np.copy(parameter_values)
        theta_plus[i] = theta_plus[i] + epsilon #increasin one parameter at a time
        #get the parameters that are changed as a dictionary again to get cost of theta_plus, theta_minus
        parameters_plus = vector_to_dictionary(theta_plus)
        J_plus[i], _ = forward_propagation_n(X, Y, parameters_plus)
        
        theta_minus = np.copy(parameter_values)
        theta_minus[i] = theta_minus[i] - epsilon
        parameters_minus = vector_to_dictionary(theta_minus)
        J_minus[i], _ = forward_propagation_n(X, Y, parameters_minus)
        
        gradapprox[i] = (J_plus[i] - J_minus[i]) / (2*epsilon) #observe it is cost per parameter increase
        
    #difference
    numerator = np.linalg.norm(grad - gradapprox)
    denominator = np.linalg.norm(grad) + np.linalg.norm(gradapprox)
    difference = numerator / denominator
    
    if print_msg:
        if difference < 2e-7:
            print("\033[92m" + "Your backward propagation works perfectly fine! difference = " +str(difference) + "\033[0m")
        else:
            print("\033[93m" + "There is a mistake in the backward propagation! difference = "+str(difference) + "\033[0m")
    return difference
```
