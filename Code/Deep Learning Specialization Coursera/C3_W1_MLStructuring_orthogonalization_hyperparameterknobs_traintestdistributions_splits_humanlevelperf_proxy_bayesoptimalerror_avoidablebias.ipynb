### Orthogonalization
    Orthogonalization is a process which we should always follow in the iterative improvement of ML Model. It literally means "One step a time". Orthogonalization is a process of tuning one hyperparameter at a time instead of tuning all at a time. This doesnt include only tuning hyperparamters like learningrate or beta and so on. It also includes decisions like adding more data,  or adding regularization or making the network bigger and so on.
    
    It basically means, using any one of the ideas that mentioned previously that one could consider to improve the model in a time rather thana applying more than one idea in a time. This helps us in guiding the iterative processs better. Note that orthogonalization is the technical term we use which refers to take one idea at a time and experiment and improve by taking another idea. But, the decision to decide which idea to take that actually work will be decide by bias varaince analysis.

One more know to adjust the orientation, another knob to adjust the trapezoidal alignment of the picture. This is an example of orthogonalization where we have one designated knob that we could tune to adjust one propery of the picture in a way to achieve the final goal of moving the image to the center of the TV.

### Single Number Evaluation Metric
    When evaluating a model, it is beneficial to use single evaluation metric.
    
    The basic startegy in ML set up is:
    


        - We quickly set up a model and train it with our initial knowledge until it performs well on training data (Like achieving almost human level performance)
            - May be we try different strategies and different set ups (like different activations, different hidden layers..) and got several classifiers, Lets say we got 2 classifiers classifier A, classifier B that performs well on training data for simplicity
        
        - Now, we want to evalaute which model between those 2 models are the best model so that we can select and further improve that selected model.
            - To know that, we usually look at their evalaution metrics.
            
        Now, Imagine we have 2 evalaution metrics like precision and recall. 
        
            Precision = TP / TP + FP = True Positive Predictiveness
            Recall = TP / TP + FN 
            
       Precision measures how many positives among the predicted positives are actually positive?
       Recall measures, of all the actual positives, how many positves that the model predicted correctly?
       
       If precision is 90%, then this means, 90% chance that it is actually positive
       If recall is 90%, than the model predicts the positive examples correctly 90% of the time.

    If we have 2 metrics, if you see, the classifier A is performing well on recall whereas classifier B is performing well on Precision. We want a model that performs well on both Precision and Recall. But here both the models have one metric that it performs well on. In such case, we cant decide which one is the best model, because both have best performance in one of the metrics.
    
    So, it is important to have one metric that can quickly tell us which is best model. In this scenario, we can use F1 scores that combines both these metrics using harmonic mean. So, since the F1 score is better for classifier A, we would select it to further undergo the iterative process.

See this satisficing metric as threshold metric. Lets see in detail below
    
### Optimizing Metric and Sastisficing Metric

    In classification, when we have two or more metrics that we would care about in an application, then we use the classification error or the classification accuracy as the optimization metric and we would use the other metrics as satisficing metric. 
    
    For eg, we have 2 metrics, classification accuracy and running time in milliseconds not greater than 100ms
    
    The model that highest accuracy and the runningtime atmost (<=) 100ms
 

 
 
    Classifer A has " " classification accuracy and the running time is ""
    Classifier B has "" classification accuracy and the running time is ""
    Classifier C has "" classification accuracy and the running time is ""
    
    If we dont have this running time constraint, we would say that classifier "C" is the best performing model as it has highest accuracy. But since we also care about running time now, we could do 2 things.
    
    1. We could create a single metric out of these 2 metrics.
        For eg, may be we do linear combination of these metrics accuracy - 1.26 * runningtime something like this
    2. May be use one single metric as optimization metric and the remaining other one (or more) as satisficing metrics.
        F
        We basically write the metric as:
            - Highest classification accuracy "subject to runningtime <= 100ms". See this satisficing metric as threshold metric.
       We look at models accuracies and we also look at running times, select all the models that has running time <= 100ms, and then select the model among those models that has best classification accuracy. We select the model when we are satisfied with the other metric.
    
    If we have N metrics we care about, we use 
        - 1 Metric as optimization metric (Single Metric)
        - Remianing N-1 Metrics as satisficing metric

### When to change the evalaution metric
    In some cases, after training the model, we might encounter a situation with the selected evaluation metric and we had to change our evaluation metric as the choosen one is not the best metric for our problem.
    
    Lets consider an example of Cat classifier. We are using this classifier to show cat pictures to the cat lovers in an application.



    
    We have trained 2 classifiers. classifer A ended with 3% classification error and the classifier B with 5% classification error. This means, classifier A mislabels the data as Cat 3% of the time and classifier B mislabels 5% of the time as a Cat eventhough the image is not a Cat
        
    We usually will decide that classifier A as the best the classifier and may be deploys it in the application.
    
    But since classsifier A has 3% error, it could show non cat images to the users 3% of the time, eventhough it is less percentage. But this classifier A is actually mislabeling pornograhic images as a Cat and showing it to the users. But the classifier B is not mislabelling pornograhic images as Cats but may be mislabelling squirrels, rats as cats. 
    
    The application users usually doesnt want to see the unpleasant pornograhic images in their applications, they tend to not like this classifier A despite the high accurate model, the users and even the company likes the classifier B than classifier A.
    
    So, in that case, you would come to a conclusion that, the metric you are trying to measure for both the classifiers is not the correct metric as the classification error is just checking whether the prediction is equal to the true class but its not filtering the pornographic images from the mispredictions.
    
    So, one thing we can do in this scenario is "changing the evaluation metric and retrain the model."


    
    We would change the evaluation metric so that is accounts for not allowing the pornogrpahic images.
    
    For eg, we would add a weight to our existing classification error metric when the image is pornographic, we would add higher weight like 10 to the error malking it bigger error indication to the model for that example, and weight 1 to the classification error, when it is not pornographc image.  To do this, you might need to relabel the examples to be pornographic and non pornographic. This way, you would add more weight to the misclassified pornographc images as Cats and the model will tend to not allow these images.
    
    There are many such tricks, we could do to change the matric as per the need of the application scenario at hand
    
    In summary, see the evaluation metric as dart center target. As per orthogonalization concept, we would need to set up the target first i.e, fix the evaluation  metric first. This is like one knob, to selec the target.
    
    And then, Worry seperately on improving the cost function(like chnaging it or improving it by adding weights or so). This is a separate knob that we will tune later on in the iterative process. No need to do worry aout the fixing the correct cost function in the initial stage itself.

    In short, when we usually change the evaluation  metric is, if the model we trained using one cost function metric + dev set is not working well on the production or real world application data, then one reason could be, what are you measuring during training(cost) is not the correct measure for your task and its time to change the metric then by tuning the knobs related to that. ### Train/Dev/Test set Distributions
    One important thing we should consider while splitting the data is, the distributions of dev set and test set.

    We know that we optimize the model hyperparameters by tuning hyperparameters using an evaluation metric on dev set. Imagine it as dart game. Dart game will have target at the center and we will try to throws arrows to finally reach the goal of hitting the target at the center.
    
    The same analgoy applies to dev set + metric
    
    Imagine optimizing the metric to the minimum(or maximum) as the hitting the target at the center. The target is evaluation metric and hitting the target at center is similar to optimize the evaluation metric so that it reaches its minimum. So we place the target i.e. we decide on the evaluation metric. And now, optimizing the metric to minimum using different model iterative tuning techniques as trying to hit the target several times with arrows until it reaches the target center.
    
    So, its like we are placing target in a position(which is getting the best possible accuracy or error) and we are trying to do is improving the model by doing this tuning experimenting process iteratively for hours, days, or even months to reach that target. Once we have achieved the performance we need, we would be ready to take that trained model for use. 
    
    But while using it, what if we change the location of our dart target and place it in another position. So, far we have trained the model to hit dart target that is placed in one position. So, the model is trained to do well only when we place the target in that position. Once the position of the dart target changed, the model no longer able to hit the target properly of this new position.

### Bayes optimal error

    A model is basically mapping input X to its output Y. We usually evaluate how the model is mapping input X to Y using metrics like classification error or MSE and so on. We improve the model iteratively to make this error minimum.
    
    Theoretically Bayes optimal error is the best possible error that any model or human or something can acheive on a task.  There is no other model or a thing that could get low error than Bayes optimal error. Bayes optimal error is the best error that can be achieved in a task. Human level performance is usually below the Bayes optimal error and sometimes, there will be a lot of gap between them. Bayes optimal error is the superior.
    
    Our goal is to first surpass the HUman Level performance. Once we surpass the human level performance, we dont know how best can we need to achieve to reach the Bayes optimal error. Even though we know that is the best possible performance that could be achieved on that task, we actually dont know what is the values of that Bayes optimal error. So, after surpassing human level error, we try to keep improving our model, but the techniques that we are supposed to use like adding more data, increasing model architecture is not clear because we have no clear vision for comparison after the known human level performance.
    
 
 
    

### Human Level Performance as proxy for Bayes Optimal Error
    In some tasks, Humans are really really good like naturally perceived tasks like image detection, speech recognition, medical imaging diagnosis.
   
     In such cases, we could consider human level performance as proxy for Bayes optimal error. We basically telling than human level performance is the estimate of Bayes optimal error. This means, Bayes optimal error and human level performance is somewhat closer and we try to improve the model to reach that proxy human level performance.
    
    


    For example, if we have a scenario where the task is diagnoising a medical image like hand xray. We have the different human level performances.

    1. we have diagnoisis from a medical student and the classification error is 3%

    3. we have diagnoisis from a experienced doctor and the classification error is 0.7%
    4. we have diagnoisis from group of experienced doctors debates and the classification error is 0.5%
    
    If we consider human level performance as proxy for Bayes optimal error, then you would select 4th option "Diagnoisis from group of experienced doctors debates that achieved 0.5 classification error" as the human level performance.
    
    Selecting which diagnosis as human level performance depends on the need of the task. If you are just developing a model for your research paper and your goal is not reach the proxy human level performance but to reach one human performance, you may consider human level performance as option 1 or 2. It all depends on what your goal is. Whether you need your make your model to perform accurately in  medical diagnosis, you may use 4th option as proxy.
    
    

### Avoidable Bias

    We talked about how you want your learning algorithm to do well on the training set but sometimes you don't actually want to do too well and knowing what human level performance is, can tell you exactly how well but not too well you want your algorithm to do on the training set. Let see what it means. 
    
    We have used Cat classification a lot and given a picture, let's say humans have near-perfect accuracy so the human level error is 1%. In that case, if your learning algorithm achieves 8% training error and 10% dev error, then maybe you want it to do better on the training set. 
    
    So the fact that there's a huge gap between how well your algorithm does on your training set vs humans do shows that your algorithm isn't even fitting the training set well. So in terms of tools to reduce bias or variance, in this case we would focus on reducing bias. So you want to do things like train a bigger neural network or run training set longer, just try to do better on the training set. 
    
        But now let's look at the same training error and dev error and imagine that human level performance was not 1%. Think of may be you know in a different application or maybe on a different data set, let's say that human level error is actually 7.5%. Maybe the images in your data set are so blurry that even humans can't tell whether there's a cat in this picture. This example is maybe slightly contrived because humans are actually very good at looking at pictures and telling if there's a cat in it or not. But for the sake of this example, let's say your data sets images are so blurry or so low resolution that even humans get 7.5% error. In this case, even though your training error and dev error are the same as the other example, you see that maybe you're actually doing just fine on the training set.It's doing only a little bit worse than human level performance. And in this second example, you would maybe want to focus on reducing the variance in your learning algorithm. So you might try regularization to try to bring your dev error closer to your training error for example. 
    


    
    So in the earlier course's discussion on bias and variance, we were mainly assuming that there were tasks where Bayes error is nearly 0. So to explain what just happened here, for our Cat classification example, think of human level error as a proxy or as a estimate for Bayes error or for Bayes optimal error. And for computer vision tasks, this is a pretty reasonable proxy because humans are actually very good at computer vision and so whatever a human can do is maybe not too far from Bayes error. By definition, human level error is worse than Bayes error because nothing could be better than Bayes error but human level error might not be too far from Bayes error. 
    
    So the surprising thing we saw here is that depending on what human level error is or really this is really approximately Bayes error or so we assume it to be, but depending on what we think is achievable, with the same training error and dev error in these two cases, we decided to focus on bias reduction tactics or on variance reduction tactics. 
    
    And what happened is in the example on the left, 8% training error is really high when you think you could get it down to 1% and so bias reduction tactics could help you do that. Whereas in the example on the right, if you think that Bayes error is 7.5% and here we're using human level error as an estimate or as a proxy for Bayes error, but you think that Bayes error is close to 7.5% then you know there's not that much headroom for reducing your training error further down. You don't really want it to be that much better than 7.5% because you could achieve that only by maybe starting to over fit the training set, and instead, there's much more room for improvement in terms of taking this 2% gap and trying to reduce that by using variance reduction techniques such as regularization or maybe getting more training data. 
    
    So to give these things a couple of names, this is not widely used terminology but I found this useful terminology and a useful way of thinking about it, which is I'm going to call the difference between Bayes error or approximation of Bayes error and the training error to be the "avoidable bias". So what you want is to maybe keep improving your training performance until you get down to Bayes error but you don't actually want to do better than Bayes error. You can't actually do better than Bayes error unless you're overfitting. And the difference between your training area and the dev error, there's a measure still of the variance problem of your algorithm. And the term avoidable bias acknowledges that "there's some bias or some minimum level of error that you just cannot get below" which is that if Bayes error is 7.5%, you don't actually want to get below that level of error.

### Improving your model performance (Guidelines Summary)

    You've heard about orthogonalization, how to set up your dev and test sets, human-level performance as a proxy for Bayes error and how to estimate your avoidable bias and variance. Let's pull it all together into a set of guidelines to how to improve the performance of your learning algorithm. 
    
    So, I think getting a supervised learning algorithm to work well means fundamentally hoping or assuming they can do two things. 
    
   - First, is that you can fit the training set pretty well, and you can think of this as roughly saying that you can achieve low avoidable bias. 
   - And the second thing you're assuming you can do well, is that doing well on the training set generalizes pretty well to the dev set or the test set, and this is sort of saying that variance is not too bad. 
   
    And in the spirit of orthogonalization, what you see is that there's a certain set of knobs you can use to fix avoidable bias issues, such as training a bigger network or training longer, and as a separate set of things you could use to address variance problems, such as regularization or getting more training data. 
   
    So, to summarize up the process we've seen so far, If you want to improve the performance of your machine learning system , I would recommend
    
   - Looking at the difference between your training error and your proxy for Bayes error and just gives you a sense of the avoidable bias. In other words, just how much better do you think you should be trying to do on your training set.
    
   - And then look at the difference between your dev error and your training error as an estimate of how much of a variance problem you have. In other words, how much harder you should be working to make your performance generalized from the training set to the dev set that it wasn't trained on explicitly.



    So, to whatever extent you want to try to reduce avoidable bias, I would try to apply tactics like train a bigger model. So, you can just do better on your training sets or train longer, use a better optimization algorithm, such as ADAM, momentum or RMSprop, or one other thing you could try is to just find a better neural network architecture or better set of hyperparameters, and this could include everything from changing the activation function to changing the number of layers or hidden units. Although if you do that, it would be in the direction of increasing the model size to trying out other models or other model architectures, such as recurrent neural networks and convolutional neural networks, which we'll see in later courses. Whether or not a new neural network architecture will fit your training set better is sometimes hard to tell in advance, but sometimes you can get much better results with a better architecture. 
    


    Next to the extent that you find out variance is a problem, some of the many techniques you could try then includes the following: you can try to get more data because getting more data to train on could help you generalize better to dev set data that your algorithm didn't see, you could try regularization. So, this includes things like L2 regularization or dropout or data augmentation, which we talked about in the previous course, or once again, you can also try various neural network architecture/hyperparameters search to see if that can help you find a neural network architecture that is better suited for your problem. 
    
    I think that this notion of bias or avoidable bias and variance is one of those things that's easily learnt but tough to master. And if you're able to systematically apply the concepts from this week's video, you actually will be much more efficient and much more systematic and much more strategic than a lot of machine learning teams in terms of how to systematically go about improving the performance of your machine learning system.



     In summary, what orthogonalization, bias varaince analysis telling us to do is:
         - First fit the training set well i.e make the model to reach better performance like single human level performance. Until then understand it has bias problem and try to tune any known konbs to make it better
         - One we reaches the better performance on training set, evaluate on dev set to know if it has variance problem. Tune another knobs based on your bias varaince analysis, tune one knob to fix variance related issues and then go for another knobs 
         - Then once you are satisfied with the dev set performance, may be test the model on test set. If you see that model is not performing well on test set, this means you overtuned the dev set.(like overfitting the dev set). Then you may tune tune the knob of increasing the dev set
         - Once you are satisfied with test set, you deployed the model and if its not performing well on the test set, then you may come to the thought that may be the metric we are measuring is not the correct metric for our task at hand. SO may we tune the knob of changing the metric or may be changing the dev set to include more diverse examples that resembles real world(production data) and retrain 
         
       Note that early stopping is a technique to avoid overfitting. But it doesnt support orthogonalization, As we stop the training early, if the model has variance problem or bias problem, if we turn one knob, we ought to change another knob as well. Usually not reaaly great to use early stopping
