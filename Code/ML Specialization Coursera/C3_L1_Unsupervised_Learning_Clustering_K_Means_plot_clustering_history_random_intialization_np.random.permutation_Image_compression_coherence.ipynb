### K-Means Clustering Algorithm

    Intialise the K cluster centers or centroids randomly (μ1, μ2, ..μk)
    Repeat until convergence,
        1. Assignment Step
            For each example(i=1,..m), Assign Ci to be the index of the cluster centroid that is closest(euclidean distance) to each data point i. So Ci would be any nymber between (1,2.,, K)
            This is mathematically done by finding cluster index k that minimised its distance from point xi i.e
                              min_k ||xi - μk||^2 and assign k to Ci
        2. Update Step
            Update the cluster centroids μ1, μ2, ..μk by
            Calculating the Mean of the data points having same Ci. This moves the centroids closer to the center of their respective clusters.
        
            where "||x - μ||" called L2-Norm which represents the Euclidean distance between 2 points (here a data point and its assigned cluster's centroid).
            
    ________________________________________________________________________________________________________________________
     
      K-Means algorithm is essentially is optimizing an cost function inorder to achieve convergence. In the earlier courses, courses supervised and advance learning algorithms, you saw a lot of supervised learning algorithms as taking training set posing a cost function. And then using grading descent or some other algorithms to optimize that cost function. It turns out that the k-means algorithm that you saw above is also optimizing a specific cost function. Although the optimization algorithm that it uses to optimize that is not gradient descent.
      
     K-Means optimization Objective:
     -------------------------------
         The objective of the K-means algorithm is to find the cluster assignments and the cluster centroids that minimize this cost function 'J.' i.e min J(c1..cm, μ1..μk)
         
         where the cost function is J(c1..cm, μ1..μk) =  1/m Σ ||xi - μ_ci||^2
         
     The cost function for K-means is the average squared distance between every training example xi and the location of the cluster centroid  μ_ci to which this training example xi has been assigned
         
         Objective is to min J(c1..cm, μ1..μk) = min 1/m Σ ||xi - μ_ci||^2
     
     And it turns out that what the K means algorithm is doing is, trying to find assignments of points to clusters centroids(1st step) as well as find locations of clusters centroids(2nd step) at every step that minimizes the squared distance.
        
    Convergence is when, there is no change in assignments of all the points to its cluster centroid, when the cluster centroid positions are not changing, when the reduction in cost is diminishing or very very small, stop the algorithm and declare convergence
    
     Note: The cost function is sometimes also called "distortion function" or "interia"
     Note: Centroids acts as representative of the clusters
    Note: The dimensions of the centroids vector will be same as the dimensions of input data i.e (m, n) where as each centroid is of shape (1, n)
    Note: Cluster centroids with no points belongs to it can be just discarded and we end up with K-1 clusters. But if we really need K-clusters, then an alternative would be to just randomly reinitialize that cluster centroid and hope that it gets assigned at least some points next time round. But it's actually more common when running K-means to just eliminate a cluster if no points are assigned to it
    
##### Drawbacks of K-Means clustering algorithm
    - K-Means work well only for circular clusters
        It won't perform well for the clusters of different shapes like spiral or elipses or circles with in circles or clusters with different denisties.
    - K-means sometimes applies to data with well separated clusters as well as data with not well separated clusters cuz of its simplicity and efficiency 
    - Choice of K afftects the clustering process. Higher value of K has high variance. Also choosing higher k, eg k = m, then almost all points will have a closer cluster centroid and the cost J will be zero. Lower value of K has high bias. For eg when k =1, all points belongs to same cluster and it leads to higher cost J
    - Random initialization also affects the clusters. Each random initialization leads to different clusters

##### Coherence and Cohesive Clusters
    Cohesive clusters and coherent clusters are concepts related to clustering algorithms used in data analysis and machine learning. They aim to group similar data points together based on certain features or characteristics. While the terms are sometimes used interchangeably, they have slightly different connotations

    Cohesive Clusters:
    ------------------
    Cohesive clusters refer to clusters where data points within each cluster are tightly packed together and have a strong intra-cluster similarity. In other words, the data points within a cohesive cluster are more similar to each other compared to the data points in other clusters. The goal of cohesive clustering is to minimize the within-cluster variance or distance, ensuring that the members of a cluster are closely related.

    Coherent Clusters:
    ------------------
    Coherent clusters, on the other hand, refer to clusters where data points have a high similarity to a central point or centroid, often representing the center of the cluster. In this case, the data points may not be as tightly packed as in cohesive clusters, but they are still relatively close to the cluster's central point. Coherent clustering aims to maximize the similarity of data points to the centroid of their assigned cluster.

    It's important to note that the degree of cohesion and coherence in clusters depends on the chosen clustering algorithm and the specific distance metric used to measure the similarity between data points. Different clustering algorithms, such as K-means, DBSCAN, hierarchical clustering, etc., have their own ways of creating clusters based on different criteria and objectives.
    
##### k - means aims to coherence or cohesive clusters?
    K-means aims to create cohesive clusters. The main objective of the K-means algorithm is to partition a given dataset into K clusters, where each data point belongs to the cluster with the nearest mean (centroid). The "mean" here refers to the centroid of a cluster, which is the arithmetic average of all data points in that cluster.

    The algorithm works iteratively to minimize the within-cluster sum of squared distances, also known as the "inertia" or "sum of squares error." It tries to find cluster centers (centroids) that minimize the distance between the data points within a cluster and their respective centroid. This process of minimizing the within-cluster variance effectively leads to cohesive clusters, where data points within each cluster are closely related to one another.

    In summary, K-means aims to create cohesive clusters by minimizing the within-cluster variance and grouping data points that are closer to each other based on the chosen distance metric (usually Euclidean distance). However, it's worth noting that the quality of the resulting clusters can be sensitive to the initial choice of centroids and the number of clusters (K) specified in the algorithm.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def find_closest_centroid(X, centroids):
    """
    Computes the centroid memberships for every example
    
    Args:
        X (ndarray): (m, n) Input values      
        centroids (ndarray): k centroids
    
    Returns:
        C_idx (array_like): (m,) closest centroids
    
    """
    m, n = X.shape
    C_idx = np.zeros(len(X), dtype=int) 
    for i in range(m):
        clusters_distances = np.zeros(len(centroids))
        for j in range(len(centroids)):
            #manual
#             euclidean_distance = np.sqrt(np.sum(X[i] - centroids[j]**2))
#             sqaured_euclidean_distance = np.square(euclidean_distance)
#             clusters_distances[j] = sqaured_euclidean_distance
            #calculates euclidean distance
            euclidean_distance = np.linalg.norm(X[i] - centroids[j])
            clusters_distances[j] = euclidean_distance ** 2
#         print(clusters_distances.shape) (3,)
        C_idx[i] = np.argmin(clusters_distances, axis=0)
    return C_idx
```

##### 1.2 Computing centroid means
    Given assignments of every point to a centroid, the second phase of the algorithm recomputes, for each centroid, the mean of the points that were assigned to it.


    Exercise 2
    ----------
    Please complete the update_centroids below to recompute the value for each centroid

    Specifically, for every centroid μk, we set μk = 1 / |C_k| * Σ xi where i ∈ Ck


    where C_k is the set of examples that are assigned to centroid k, |C_k| is the number of examples in the set C_k


    Concretely, if two examples say x_3 and  x_5 are assigned to centroid k=2, then you should update μ2 = 1/2 (x_3 + x_5)


```python
def update_centroids(X, K, C):
    """
    Returns the new centroids by computing the means of the 
    data points assigned to each centroid.
    
    Args:
        X (ndarray):   (m, n) Data points
        K (int):       number of centroids
        C (ndarray):   (m,) Array containing index of closest centroid for each 
                       example in X. Concretely, C[i] contains the index of 
                       the centroid closest to example i
    
    Returns:
        new_centroids (ndarray): (K, n) New centroids computed
    """
    new_centroids = np.zeros((K, X.shape[1])) 
    for j in range(K):
        #get indexes of all the points using C
        sample_indexes = np.where(C==j)[0] #X[C == j]  also works
        new_centroids[j] = np.mean(X[sample_indexes], axis=0)
    return new_centroids
```

    After you have completed the two functions (find_closest_centroids and compute_centroids) above, the next step is to run the K-means algorithm on a toy 2D dataset to help you understand how K-means works.

    Notice that the code calls the two functions you implemented in a loop.
    When you run the code below, it will produce a visualization that steps through the progress of the algorithm at each iteration.

```python
def draw_line(new_centriods, previous_centroids):
    p1 = new_centriods #(x1, x2)
    p2 = previous_centroids #(x1, x2)
    plt.plot([p1[0], p2[0]], [p1[1], p2[1]], "-k", lw=1)
    
def plot_data(X, C):
    plt.scatter(X[:, 0], X[:, 1], c=C)
    
def plot_progress_kMeans(X, C_X, new_centroids, previous_centroids):
    #first plot the data with their assignment to previous clsuters before updating
    plot_data(X, C_X)
    
    #now plot the updated clusters centroids
    plt.scatter(new_centroids[:, 0], new_centroids[:, 1], c="k", marker="x", lw=3)
    
    #plot the history of previous centroids updating to new centroids
    for i in range(len(previous_centroids)):
        draw_line(new_centroids[i], previous_centroids[i])

def run_kMeans(X, K, initial_centroids, max_iters=10, plot_progress=True):
    m, n = X.shape
    centroids = initial_centroids
    previous_centroids = centroids #for plotting the path of K-means centroids
    C_X = np.zeros(m,)
    for i in range(max_iters):
        #Output progress
        print("K-Means iteration %d/%d" % (i, max_iters-1))
        #Assigning
        C_X = find_closest_centroid(X, centroids)
        if plot_progress:
            plot_progress_kMeans(X, C_X, centroids, previous_centroids)    
        previous_centroids = centroids
        #update the centroids
        centroids = update_centroids(X, K, C_X)
        plt.title("Iteration number %d" %i)
    plt.show() #showing at the end of for loop to display only the last plot
    return centroids, C_X
        
initial_centroids = np.array([[3, 3], [6, 2], [8, 5]])
K = 3
centroids, C_idx = run_kMeans(X, K, initial_centroids, max_iters=10)   
    
        
```

#### 3 - Random initialization
    The initial assignments of centroids for the example dataset was designed so that you will see the same figure as in above Figure. In practice, a good strategy for initializing the centroids is to select random examples from the training set.

    In this part of the exercise, you should understand how the function kMeans_init_centroids is implemented.

    The code first randomly shuffles the indices of the examples (using np.random.permutation()).
    Then, it selects the first K examples based on the random permutation of the indices.
    This allows the examples to be selected at random without the risk of selecting the same example twice.

```python
def kMeans_init_centroids(X, K):
    """
    This function initializes K centroids that are to be 
    used in K-Means on the dataset X
    
    Args:
        X (ndarray): Data points 
        K (int):     number of centroids/clusters
    
    Returns:
        centroids (ndarray): Initialized centroids
    """
    rand_idx = np.random.permutation(X.shape[0]) # Randomly reorder the indices of examples
    centroids = X[rand_idx[:K]] # Take the first K examples as centroids
    
    return centroids
```

#### Notes on Image Compressions and color representations:
       Image compression is a process of reducing the file size of an image while trying to maintain its visual quality as much as possible. There are two main types of image compression: lossless compression and lossy compression.

    Lossless Compression:
    _____________________
        Lossless compression aims to reduce file size without losing any image data. The basic idea is to find patterns and redundancies in the image and represent them more efficiently. The procedure typically involves these steps:
        
        a. Pixel Prediction: 
        ---------------------
            The compression algorithm predicts the value of each pixel based on its neighboring pixels. It tries to find patterns or similarities in neighboring pixel values.

        b. Encoding: 
        -------------
            The difference between the predicted pixel values and the actual pixel values is computed. This difference, also known as the residual, is then encoded using various techniques like Huffman coding or Run-Length Encoding.

        c. Entropy Coding:
        -------------------
            Entropy coding further reduces the size by assigning shorter codes to frequently occurring patterns and longer codes to less frequent ones.

        d. Decoding: 
        ------------
            When you want to view the compressed image, the reverse process is applied. The compressed data is decoded to recreate the original pixel values.

        The result of lossless compression is an image that is identical to the original, but with a smaller file size. However, lossless compression may not achieve as high compression ratios as lossy compression.

    Lossy Compression:
    __________________
    Lossy compression sacrifices some image details to achieve higher compression ratios. The basic steps of lossy compression are:
        a. Transform Coding: 
        ---------------------
            The image is transformed into a different domain, such as the frequency domain using techniques like Discrete Cosine Transform (DCT). This conversion separates the image into different frequency components.

        b. Quantization:
        -----------------
            The transformed data is reduced in precision or quantized. This means that some less essential information is discarded, resulting in a loss of quality.

        c. Encoding:
        ------------
        The quantized data is then further compressed using techniques like Huffman coding or arithmetic coding to reduce the file size.

        d. Decoding:
        -------------
        When viewing the image, the compressed data is decoded, and the image is reconstructed, but with some loss of detail due to quantization.

        The advantage of lossy compression is that it can achieve much higher compression ratios than lossless compression. However, as the name suggests, it results in some loss of image quality. The challenge is to find the right balance between compression and acceptable image degradation.

    Different image compression algorithms use various combinations of these techniques to achieve different compression levels and image quality trade-offs. Popular image compression formats include JPEG (lossy) and PNG (lossless).

##### Color representation
    The color representation of images in bits depends on the color depth or the number of bits assigned to each color channel (red, green, and blue) of a pixel. 
    
    The color depth determines the number of colors that can be represented in an image.

    Color Depths:
    -------------
    1-bit color:
    With 1-bit color depth, each pixel can be either black or white. This means each pixel is represented by 1 bit, where 0 represents black and 1 represents white.

    8-bit color:
    With 8-bit color depth, each color channel (red, green, and blue) can have 256 levels of intensity, ranging from 0 to 255. This allows for a total of 256^3 (16,777,216) possible colors. Each pixel in the image is represented by 24 bits (8 bits for each color channel).

    16-bit color:
    With 16-bit color depth, each color channel has 65,536 levels of intensity (2^16). This allows for more color variations and smoother gradients. Each pixel is represented by 48 bits (16 bits for each color channel).

    24-bit color (True Color):
    With 24-bit color depth, each color channel has 256 levels of intensity (8 bits), resulting in a total of 16,777,216 possible colors. This is the most common color representation for digital images. Each pixel is represented by 24 bits (8 bits for each color channel).

    32-bit color (True Color with Alpha channel):
    In some cases, images use an additional channel called the alpha channel to represent transparency information. This is commonly used in graphics and image editing. With 32-bit color depth, each pixel is represented by 32 bits (8 bits for each color channel and 8 bits for the alpha channel).

    Higher color depths, such as 48-bit or 64-bit, are used in specialized applications and image processing tasks that require very high color precision. However, for most general purposes, 24-bit color depth (True Color) is sufficient to represent images with good color fidelity and visual quality.

    Images can be represented using different color models or color spaces. Each color model has its own way of describing and representing colors. Here are some common color representations used in images:

    RGB (Red, Green, Blue):
    ---------------------
    RGB is the most widely used color model for digital images. In this model, each pixel is represented by three color channels: red, green, and blue. The intensity of each channel ranges from 0 to 255, where 0 is no intensity (absence of the color) and 255 is the highest intensity (full color).
    For example, if a pixel has RGB values of (255, 0, 0), it means the pixel is displaying pure red, and if the values are (0, 255, 0), it displays pure green, and so on. Combining different intensities of these three primary colors creates a wide range of colors.

    CMYK (Cyan, Magenta, Yellow, Key/Black):
    ----------------------------------------
    CMYK is commonly used in printing. It represents colors by using four channels: cyan, magenta, yellow, and black. The combination of these colors can create a wide range of hues, and black (also called the "key" color) is added for better contrast and details in printed images.

    HSL (Hue, Saturation, Lightness):
    --------------------------------
    HSL is a cylindrical representation of colors. It uses three components: hue, saturation, and lightness. Hue represents the type of color (e.g., red, blue, green), saturation controls the intensity or purity of the color, and lightness determines how bright or dark the color is.

    HSV (Hue, Saturation, Value):
    ------------------------------
    HSV is similar to HSL but uses "value" instead of "lightness." The value component represents the brightness of the color instead of its perceived lightness.

    YUV (Luma, Chrominance):
    -----------------------
    YUV separates the brightness information (luma) from the color information (chrominance). Y represents the luminance or brightness component, while U and V represent the color components. This representation is commonly used in video compression.

    Lab (CIELAB):
    -------------
    Lab is a color model designed to be perceptually uniform, meaning the Euclidean distance between two colors in Lab space roughly corresponds to the perceived difference in color by the human eye. It has three components: L* for lightness, a* for the green-red axis, and b* for the blue-yellow axis.

    These are just a few examples of color representations used in digital images. Each color space has its advantages and is suited for specific applications, such as photography, printing, video, or image processing. Converting between color spaces is common when performing various image processing tasks.

#### 4.3 Compress the image
    After finding the top K = 16 colors to represent the image, you can now assign each pixel position(each pixel in the image) to its closest centroid using the find_closest_centroids function.

    This allows you to represent the original image using the centroid assignments(which are indexes C_X_idx) of each pixel.

    Notice that you have significantly reduced the number of bits that are required to describe the image.
    
    The original image required 24 bits for each one of the 128 x 128 pixel locations, resulting in total size of 128 x 128 x 24 = 393,216 bits. 
    
    Note: Each pixel has 3 color channels (rgb) and each color channel ranges from 0 to 255 intensities or values. Each of this intensity will be represented in 8 bits.
    
    The new representation requires some overhead storage in form of a dictionary of 16 colors, each of which require 24 bits, but the image itself then only requires 4 bits per pixel location as we are gonna just store the index of the colors (stored in dictionary) in the image pixel location instead of storing entire 24bits of these 16 colors.
    
    The final number of bits used is therefore 16 x 24 + 128 x 128 x 4 = 65,920 bits which corresponds to compressing the original image by about a factor of 6.

##### Detailed explanation of compression bits
    In the original image, each pixel was represented using 24 bits, where each bit is either 0 or 1. This means there were 2^24 = 16,777,216 or 16 million possible colors (24 bits can represent 2^24 different values). However, in the compressed version, the image is reduced to only 16 colors. Instead of directly encoding each pixel with 24 bits, we now use an indexed approach.

    Here's how the compression works:

    Dictionary Creation: First, a dictionary is created to store the 16 unique colors that will be used in the compressed image. Each color in the dictionary is represented using 24 bits (same as the original image).

    Indexing: Instead of directly storing the color value for each pixel, we now store an index that refers to a color in the dictionary. Since we have only 16 colors in the dictionary, we can represent each index using only 4 bits (2^4 = 16). This is the key to reducing the size of the image significantly.

    Let's break down the storage requirements:

    Dictionary: The dictionary has 16 colors, and each color requires 24 bits. So, the total bits used for the dictionary is 16 x 24 = 384 bits. Now C_X_idx here will be treated as this color dictionary

    Compressed Image: The compressed image contains the 4-bit index for each pixel location. Since the image is 128 x 128 pixels, the total bits used for the compressed image is 128 x 128 x 4 = 65,536 bits.

    Now, let's calculate the total number of bits required for the compressed image:

    Total bits = Dictionary bits + Compressed Image bits
    Total bits = 384 bits + 65,536 bits
    Total bits = 65,920 bits

    As you can see, the compressed image requires 65,920 bits compared to the original image's 393,216 bits. This corresponds to a compression ratio of approximately 6 (393,216 / 65,920 ≈ 5.95), which means the image has been compressed by about a factor of 6.

    Binary representation:
    2^24: In binary representation, 2^24 is equal to 16,777,216. This means that 24 bits can represent 16,777,216 different values, ranging from 0 to 16,777,215.

    2^4: In binary representation, 2^4 is equal to 16. This means that 4 bits can represent 16 different values, ranging from 0 to 15.


    Finally, you can view the effects of the compression by reconstructing the image based only on the centroid assignments.

    Specifically, you can replace each pixel location with the mean of the centroid assigned to it.
